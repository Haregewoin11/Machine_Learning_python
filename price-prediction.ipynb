```python
#converting the text value to numerical value
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

houseData['street'] = encoder.fit_transform(houseData['street'])
houseData['city'] = encoder.fit_transform(houseData['city'])
houseData['statezip'] = encoder.fit_transform(houseData['statezip'])
houseData['country'] = encoder.fit_transform(houseData['country'])


```

### Feature extraction

```python
houseData['date'] = pd.to_datetime(houseData['date'])
houseData['date'] = houseData['date'].dt.strftime('%Y%m%d').astype(int)
# covert the date of the csv data to pandas datetime format
```

```python
X = houseData.drop(['date','street', 'price'], axis=1)
y = houseData['price']
# ['bedrooms','bathrooms','sqft_living', 'sqft_lot','floors','waterfront','view','condition']

X
```

### Data splitting

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.20,random_state=2)
X_train
```

### Train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train,Y_train);
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import r2_score
```

```python
# model performance  
r2 = r2_score(Y_test, y_pred)

print("R-squared:", r2)
```

```python
from sklearn.linear_model import Ridge
```

```python
model2=Ridge(alpha=1.0)
model2.fit(X_train,Y_train);
```

```python
y_pred2=model2.predict(X_test)
```

```python
# Evaluate model performance  
r2 = r2_score(Y_test, y_pred2)

print("R-squared:", r2)
```
